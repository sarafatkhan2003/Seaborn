**SESSION 1: Seaborn Basics & Setup**

**What is Seaborn?**

Seaborn is a Python data visualization library built on top of Matplotlib.

- Matplotlib = low-level (more control, more code)
- Seaborn = high-level (less code, better default visuals)

Why Data Analysts Prefer Seaborn

- Cleaner charts by default
- Built-in statistical plots
- Works directly with Pandas DataFrames
- Faster insight generation

**Seaborn vs Matplotlib**

| Feature           | Matplotlib | Seaborn      |
| ----------------- | ---------- | ------------ |
| Level             | Low-level  | High-level   |
| Default look      | Plain      | Professional |
| Statistical plots | ❌          | ✅            |
| Pandas friendly   | ⚠️         | ✅            |

**SESSION 2: Distribution Analysis**

**What is Distribution?**

A distribution shows:

- How often values occur
- Where most data lies
- Whether data is skewed
- Presence of outliers

Real-World Examples

- Salary distribution in a company
- Exam score distribution
- Order amount distribution

**Histogram**

What a Histogram Shows

- X-axis → value range
- Y-axis → frequency (count)

```python
# Basic Histogram
sns.histplot(df['total_bill'])
plt.show()
```

```python
# Control Number of Bins
sns.histplot(df['total_bill'], bins=10)
plt.show()
```

When to Use Histogram

- Numerical data
- Understand spread
- First look at data

```python
# Load dataset
tips = sns.load_dataset("tips")

```

```python
# Histogram with 5 bins
plt.figure()
sns.histplot(tips["tip"], bins=5)
plt.title("Tip Distribution (5 bins)")
plt.show()
```

```python
# Histogram with 15 bins
plt.figure()
sns.histplot(tips["tip"], bins=15)
plt.title("Tip Distribution (15 bins)")
plt.show()
```

**KDE Plot**

```python
sns.kdeplot(df['total_bill'])
plt.show()
```

```python
# Histogram + KDE Together
sns.histplot(df['total_bill'], kde=True)
plt.show()
```

**Practice**

```python
# load data set
tips = sns.load_dataset("tips")
```

```python
# KDE plot for tip
plt.figure()
sns.kdeplot(tips["tip"], fill=True)
plt.title("KDE Plot of Tip")
plt.show()
```

```python
# Histogram + KDE for total_bill
plt.figure()
sns.histplot(tips["total_bill"], kde=True)
plt.title("Histogram + KDE of Total Bill")
plt.show()
```

**Box Plot**

What Box Plot Shows

- Median
- Quartiles (Q1, Q3)
- Outliers

```python
# Basic Box Plot
sns.boxplot(x=df['total_bill'])
plt.show()
```

```python
# Box Plot by Category
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

**practice**

```python
# Detect outliers in tip
plt.figure()
sns.boxplot(x=tips["tip"])
plt.title("Outliers in Tip")
plt.show()
```

```python
# Compare total bill across days
plt.figure()
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Total Bill Comparison Across Days")
plt.show()
```

**Violin Plot**

What is a Violin Plot?

Combination of:

- Box plot
- KDE
- Shows shape + summary together.

```python
# Basic Violin Plot
sns.violinplot(x=df['total_bill'])
plt.show()
```

```python
# Violin Plot by Category
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()
```

**INTERVIEW QUESTIONS**

(1) What is a distribution in data analysis?
- It shows how data values are spread or occur across different ranges.

(2) Difference between histogram and KDE?
- Histogram shows frequency using bars, while KDE shows a smooth probability curve.

(3) When would you use a box plot?
- To compare distributions and identify medians, spread, and outliers quickly.

(4) What are outliers?
- Outliers are extreme values that differ significantly from most data points.

(5) How do you detect outliers visually?
- Using box plots or scatter plots to spot points far from the main data.

(6) Can KDE replace histogram?
- KDE can complement but not fully replace histograms because it hides exact counts.

(7) When is violin plot preferred over box plot?
- When you want to see both distribution shape and summary statistics together.

Strong Interview Answer Example

- I usually start EDA with histograms and KDE to understand distribution, then use box plots to detect outliers and compare groups.

**SESSION 3: Relationship Analysis**

**What is Relationship Analysis?**

Relationship analysis answers questions like:

- Does X affect Y?
- As X increases, what happens to Y?
- Do different groups behave differently?

**Scatter Plot**

What Scatter Plot Shows

- Relationship between two numerical variables
- Direction (positive/negative)
- Strength (weak/strong)

```python
# Basic Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()
```

How to Read It

- Upward trend → positive relationship
- Downward trend → negative relationship
- No pattern → weak/no relationship

**Practice**

```python
plt.figure()
sns.scatterplot(x = 'total_bill', y = 'tip',data = tips)
plt.title('total bill vs tip')
plt.show()
```

**Adding Categorical Segmentation(Hue)**

Why Hue is Powerful

- It allows comparison across categories in one chart.

```python
# Scatter Plot with Hue
sns.scatterplot(
    x = 'total_bill',
    y = 'tip',
    hue = 'sex',
    data = df
)

plt.show()
```

```python
# Hue with Style & Size
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='day',
    style='time',
    size='size',
    data=df
)
plt.show()
```

**practice**

```python
# Compare tips by day
plt.figure()
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Tip Comparison by Day")
plt.show()
```

**Line Plot**

When to Use Line Plot

- Time series
- Ordered numeric variables
- Trends

```python
# Basic Line Plot
sns.lineplot(x='size', y='tip', data=df)
plt.show()
```

```python
# Line Plot with Hue
sns.lineplot(x='size', y='tip', hue='sex', data=df)
plt.show()
```

**Practice**

```python
# Average tip vs party size, trend by gender
plt.figure()
sns.lineplot(x="size", y="tip", hue="sex", data=tips, estimator="mean", marker="o")
plt.title("Average Tip vs Party Size (by Gender)")
plt.show()
```

**Interpreting Relationships**

Correlation ≠ Causation

- Just because X increases with Y:
- Does NOT mean X causes Y

**SESSION 4: Pairplot & Heatmap (Correlation Analysis)**

**Pairplot (Fastest EDA Tool)**

What is a Pairplot?

A pairplot:

- Plots every numerical column vs every other
- Shows distributions on the diagonal
- Helps detect relationships quickly

```python
# Basic Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Pairplot with Category (Hue)
sns.pairplot(df, hue='sex')
plt.show()
```

**Practice**

```python
# Pairplot with hue = day
sns.pairplot(tips, hue="day")
plt.show()
```

**Correlation (Core Analytics Concept)**

What is Correlation?

Correlation measures how strongly two numerical variables move together.

- +1 → strong positive
- 0 → no relationship
- −1 → strong negative

```python
# Compute Correlation Matrix
corr = df.corr(numeric_only=True)
```

**Heatmap (Correlation Visualization)**

Why Heatmap?

- Tables are boring.
- Heatmaps make correlation visually obvious.

```python
# Basic Heatmap
sns.heatmap(corr)
plt.show()
```

```python
# Heatmap with Values
sns.heatmap(corr, annot=True)
plt.show()
```

How to Read Heatmap

- Darker color → stronger correlation
- Diagonal = 1 (ignore)
- Focus on high absolute values

**Interpreting Results**

Example Insight:

 - Total bill and tip show a strong positive correlation, indicating higher bills generally result in higher tips.
